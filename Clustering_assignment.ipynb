```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.cluster.hierarchy as sch

from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv("/content/crime_data.csv")
```

```python
data
```

```python
def norm_func(i):
  x = (i-i.min())/(i.max()-i.min())
  return x
```

```python
norm_data = norm_func(data.iloc[:,1:])
norm_data
```

```python
dendrogram = sch.dendrogram(sch.linkage(norm_data,method="average"))
```

```python
hc = AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="complete")
hc
```

```python
y_hc = hc.fit_predict(norm_data)
Clusters = pd.DataFrame(y_hc,columns=["Clusters"])
```

```python
y_hc
```

```python
data["h_clusterid"] = Clusters
```

```python
data1 = pd.read_csv("/content/crime_data.csv")
```

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
norm_data = norm_func(data.iloc[:,1:])
```

```python
norm_data
```

```python
wcss = []
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(norm_data)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("Elbow method")
plt.xlabel("N of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model = KMeans(n_clusters=4)
model.fit(norm_data)
model.labels_
```

```python
x = pd.Series(model.labels_)
data1["Clust"] = x
```

```python
data1
```

```python
data1.iloc[:,1:5].groupby(data1.Clust).mean()
```
